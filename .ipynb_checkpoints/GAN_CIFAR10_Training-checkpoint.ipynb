# 🎨 GAN on CIFAR-10: Training & Image Generation

This notebook implements a **Generative Adversarial Network (GAN)** trained on the CIFAR-10 dataset.  
We will:
1. Explore the CIFAR-10 dataset  
2. Build a **Discriminator** and a **Generator**  
3. Train the GAN with **epoch-by-epoch visual feedback**  
4. **Automatically stop training** once the Generator is good enough to fool the Discriminator  
5. Use the trained Generator to produce **10 fake images**  
6. Visualize loss curves and accuracy trends

## 1. Imports & Setup

We import everything we need upfront:  
- **TensorFlow / Keras** for building and training the neural networks  
- **NumPy** for numerical operations  
- **Matplotlib** for plotting images and charts

```python
import numpy as np
from numpy import zeros, ones
from numpy.random import randn, randint

import tensorflow as tf
from tensorflow.keras.datasets.cifar10 import load_data
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import (
    Dense, Reshape, Flatten, Conv2D, Conv2DTranspose, LeakyReLU, Dropout
)

import matplotlib.pyplot as plt
from IPython.display import clear_output

# Reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print('TensorFlow version:', tf.__version__)
print('Setup complete ✅')
```

## 2. Explore the CIFAR-10 Dataset

CIFAR-10 contains **60,000** colour images (32×32 pixels) in **10 classes**:  
`airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck`  

Let's load the data, inspect its shape, and visualize a sample grid.

```python
# CIFAR-10 class names (index 0-9)
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Load the dataset
(trainX, trainy), (testX, testy) = load_data()

print(f'Training set  : {trainX.shape}, Labels: {trainy.shape}')
print(f'Test set      : {testX.shape},  Labels: {testy.shape}')
print(f'Pixel range   : [{trainX.min()}, {trainX.max()}]')
print(f'Number of classes: {len(class_names)}')
```

```python
# ----- Plot a 5x5 grid of random training images with their labels -----
fig, axes = plt.subplots(5, 5, figsize=(10, 10))
indices = randint(0, trainX.shape[0], 25)  # pick 25 random indices

for i, ax in enumerate(axes.flatten()):
    ax.imshow(trainX[indices[i]])
    ax.set_title(class_names[trainy[indices[i]][0]], fontsize=10)
    ax.axis('off')

plt.suptitle('Sample CIFAR-10 Training Images', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

## 3. Define the Discriminator

The **Discriminator** is a binary classifier:  
- **Input**: a 32×32×3 image  
- **Output**: probability that the image is *real* (sigmoid → 0 or 1)  

Architecture highlights:  
- Two `Conv2D` layers with stride 2 to downsample the image  
- `LeakyReLU` activations (allow small gradient when inactive)  
- `Dropout` for regularization  
- Compiled with **binary cross-entropy** loss and **Adam** optimizer

```python
def define_discriminator(in_shape=(32, 32, 3)):
    """Build and compile the Discriminator model."""
    model = Sequential(name='Discriminator')

    # --- Block 1: 32x32x3 → 16x16x128 ---
    model.add(Conv2D(128, (3, 3), strides=(2, 2), padding='same', input_shape=in_shape))
    model.add(LeakyReLU(alpha=0.2))

    # --- Block 2: 16x16x128 → 8x8x128 ---
    model.add(Conv2D(128, (3, 3), strides=(2, 2), padding='same'))
    model.add(LeakyReLU(alpha=0.2))

    # --- Classifier head ---
    model.add(Flatten())          # 8*8*128 = 8192
    model.add(Dropout(0.4))       # Prevent overfitting on real images
    model.add(Dense(1, activation='sigmoid'))  # Output: probability of being real

    # Compile with binary cross-entropy (real=1 vs fake=0)
    opt = Adam(learning_rate=0.0002, beta_1=0.5)
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    return model


# Instantiate and inspect
test_discr = define_discriminator()
test_discr.summary()
```

## 4. Define the Generator

The **Generator** maps a random **latent vector** (1D noise) to a 32×32×3 image:  
- **Input**: latent vector of size `latent_dim` (e.g., 100)  
- **Output**: a synthetic 32×32×3 image with pixel values in [-1, 1] (tanh)  

Architecture highlights:  
- `Dense` layer to project the latent vector into a small spatial volume (8×8×128)  
- Two `Conv2DTranspose` layers to upsample 8→16→32  
- Final `Conv2D` with **tanh** activation to produce 3-channel output  
- **Not compiled** directly — it is trained through the combined GAN model

```python
def define_generator(latent_dim):
    """Build the Generator model (not compiled separately)."""
    model = Sequential(name='Generator')

    # --- Foundation: project & reshape latent vector → 8x8x128 ---
    n_nodes = 128 * 8 * 8  # 8192
    model.add(Dense(n_nodes, input_dim=latent_dim))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Reshape((8, 8, 128)))

    # --- Upsample: 8x8x128 → 16x16x128 ---
    model.add(Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same'))
    model.add(LeakyReLU(alpha=0.2))

    # --- Upsample: 16x16x128 → 32x32x128 ---
    model.add(Conv2DTranspose(128, (4, 4), strides=(2, 2), padding='same'))
    model.add(LeakyReLU(alpha=0.2))

    # --- Output: 32x32x128 → 32x32x3 ---
    # tanh produces values in [-1, 1] to match our scaled real images
    model.add(Conv2D(3, (8, 8), activation='tanh', padding='same'))

    return model


# Instantiate and inspect
test_gen = define_generator(100)
test_gen.summary()
```

## 5. Define the Combined GAN Model

The GAN connects **Generator → Discriminator** into one model.  
- During GAN training the Discriminator weights are **frozen**  
- Only the Generator learns to produce images that the Discriminator classifies as *real*  
- Loss: binary cross-entropy (target = 1, i.e., "fool the Discriminator")

```python
def define_gan(generator, discriminator):
    """
    Combine Generator + Discriminator.
    The Discriminator is frozen so only the Generator gets updated.
    """
    discriminator.trainable = False  # Freeze discriminator weights

    model = Sequential(name='GAN')
    model.add(generator)
    model.add(discriminator)

    opt = Adam(learning_rate=0.0002, beta_1=0.5)
    model.compile(loss='binary_crossentropy', optimizer=opt)
    return model
```

## 6. Helper Functions

| Function | Purpose |
|---|---|
| `load_real_samples()` | Load CIFAR-10 training images and scale to [-1, 1] |
| `generate_real_samples()` | Pick a random batch of real images (label = 1) |
| `generate_latent_points()` | Create random noise vectors for the Generator |
| `generate_fake_samples()` | Use the Generator to produce fake images (label = 0) |
| `show_generated_images()` | Display a grid of generated images during training |

```python
def load_real_samples():
    """
    Load CIFAR-10 training images and rescale pixel values
    from [0, 255] to [-1, 1] to match Generator's tanh output.
    """
    (trainX, _), (_, _) = load_data()
    X = trainX.astype('float32')
    X = (X - 127.5) / 127.5  # Scale to [-1, 1]
    return X


def generate_real_samples(dataset, n_samples):
    """
    Select 'n_samples' random real images from the dataset.
    Label them as 1 (real) for the Discriminator.
    """
    ix = randint(0, dataset.shape[0], n_samples)
    X = dataset[ix]
    y = ones((n_samples, 1))  # Real label = 1
    return X, y


def generate_latent_points(latent_dim, n_samples):
    """
    Generate 'n_samples' random latent vectors drawn from a
    standard normal distribution.  These serve as input to the Generator.
    """
    x_input = randn(latent_dim * n_samples)
    x_input = x_input.reshape(n_samples, latent_dim)
    return x_input


def generate_fake_samples(generator, latent_dim, n_samples):
    """
    Use the Generator to produce 'n_samples' fake images.
    Label them as 0 (fake) for the Discriminator.
    """
    x_input = generate_latent_points(latent_dim, n_samples)
    X = generator.predict(x_input, verbose=0)
    y = zeros((n_samples, 1))  # Fake label = 0
    return X, y


def show_generated_images(generator, latent_dim, n=3, epoch=None):
    """
    Generate and display a n×n grid of fake images.
    Called after each epoch to visualize training progress.
    """
    latent_points = generate_latent_points(latent_dim, n * n)
    X = generator.predict(latent_points, verbose=0)
    # Scale from [-1, 1] back to [0, 1] for display
    X = (X + 1) / 2.0

    fig, axes = plt.subplots(n, n, figsize=(6, 6))
    for i, ax in enumerate(axes.flatten()):
        ax.imshow(X[i])
        ax.axis('off')

    title = 'Generated Images'
    if epoch is not None:
        title += f' — Epoch {epoch}'
    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()


print('Helper functions defined ✅')
```

## 7. Training the GAN

### Training strategy
1. **Discriminator** is trained on a **half-batch of real images** (label=1) and a **half-batch of fake images** (label=0) — separately, as research shows this is more effective.  
2. **Generator** is then trained through the combined GAN model: it generates a full batch of fakes and wants the Discriminator to classify them as *real* (label=1).  

### Early stopping for the Generator
We monitor the Discriminator's loss on fake images (`d_loss_fake`).  
- A **high `d_loss_fake`** means the Discriminator is struggling to identify fakes → the Generator is doing a great job.  
- If the average `d_loss_fake` exceeds **0.7** for **5 consecutive epochs**, we consider the Generator has learned enough and **stop training early**.  
- Maximum cap: **100 epochs** (safety net).  

### Visual feedback
After every epoch, a **3×3 grid** of generated images is displayed so you can watch the quality improve over time.  

> ⏱️ **Training time**: Depending on your hardware, each epoch may take a few minutes. Adjust `n_epochs` or `n_batch` as needed.

```python
def train(g_model, d_model, gan_model, dataset, latent_dim,
          n_epochs=100, n_batch=128, patience=5, d_fake_loss_threshold=0.7):
    """
    Main training loop with:
      - epoch-by-epoch generated image display
      - loss & accuracy tracking
      - early stopping when Generator is good enough

    Parameters
    ----------
    g_model       : Generator model
    d_model       : Discriminator model
    gan_model     : Combined GAN model
    dataset       : real training images scaled to [-1, 1]
    latent_dim    : size of the latent (noise) vector
    n_epochs      : maximum training epochs
    n_batch       : batch size
    patience      : how many consecutive epochs d_loss_fake must exceed
                    the threshold before we stop
    d_fake_loss_threshold : the d_loss_fake level that signals the
                           Generator is fooling the Discriminator

    Returns
    -------
    history : dict with epoch-level loss and accuracy lists
    """

    bat_per_epo = int(dataset.shape[0] / n_batch)
    half_batch = int(n_batch / 2)

    # ----- Tracking containers -----
    history = {
        'd_loss_real': [], 'd_loss_fake': [], 'g_loss': [],
        'd_acc_real': [],  'd_acc_fake': []
    }
    fooling_streak = 0  # counts consecutive epochs Generator is fooling D

    for epoch in range(1, n_epochs + 1):
        # Accumulators for this epoch
        ep_d_loss_real, ep_d_loss_fake, ep_g_loss = 0, 0, 0
        ep_d_acc_real, ep_d_acc_fake = 0, 0

        for batch in range(bat_per_epo):
            # ------ Train Discriminator on REAL images ------
            X_real, y_real = generate_real_samples(dataset, half_batch)
            d_loss_r, d_acc_r = d_model.train_on_batch(X_real, y_real)

            # ------ Train Discriminator on FAKE images ------
            X_fake, y_fake = generate_fake_samples(g_model, latent_dim, half_batch)
            d_loss_f, d_acc_f = d_model.train_on_batch(X_fake, y_fake)

            # ------ Train Generator via the combined GAN model ------
            # The Generator wants the Discriminator to output 1 (real) for fakes
            X_gan = generate_latent_points(latent_dim, n_batch)
            y_gan = ones((n_batch, 1))  # target = 1 → "fool the Discriminator"
            g_loss_batch = gan_model.train_on_batch(X_gan, y_gan)

            # Accumulate
            ep_d_loss_real += d_loss_r
            ep_d_loss_fake += d_loss_f
            ep_g_loss     += g_loss_batch
            ep_d_acc_real += d_acc_r
            ep_d_acc_fake += d_acc_f

        # ----- Epoch averages -----
        avg_d_loss_real = ep_d_loss_real / bat_per_epo
        avg_d_loss_fake = ep_d_loss_fake / bat_per_epo
        avg_g_loss      = ep_g_loss      / bat_per_epo
        avg_d_acc_real  = ep_d_acc_real  / bat_per_epo
        avg_d_acc_fake  = ep_d_acc_fake  / bat_per_epo

        history['d_loss_real'].append(avg_d_loss_real)
        history['d_loss_fake'].append(avg_d_loss_fake)
        history['g_loss'].append(avg_g_loss)
        history['d_acc_real'].append(avg_d_acc_real)
        history['d_acc_fake'].append(avg_d_acc_fake)

        # ----- Print epoch summary -----
        print(f'Epoch {epoch}/{n_epochs}  |  '
              f'd_loss_real={avg_d_loss_real:.4f}  d_loss_fake={avg_d_loss_fake:.4f}  '
              f'g_loss={avg_g_loss:.4f}  |  '
              f'd_acc_real={avg_d_acc_real:.2%}  d_acc_fake={avg_d_acc_fake:.2%}')

        # ----- Show generated images after this epoch -----
        show_generated_images(g_model, latent_dim, n=3, epoch=epoch)

        # ----- Early stopping check -----
        # If d_loss_fake is high, the Discriminator is failing on fakes
        # → Generator is producing convincing images
        if avg_d_loss_fake > d_fake_loss_threshold:
            fooling_streak += 1
            print(f'  ⚡ Generator fooling streak: {fooling_streak}/{patience}')
        else:
            fooling_streak = 0  # reset streak

        if fooling_streak >= patience:
            print(f'\n🛑 Early stopping at epoch {epoch}! '
                  f'Generator has been fooling the Discriminator for '
                  f'{patience} consecutive epochs.')
            break

    # ----- Save the trained Generator -----
    model_path = 'cifar_gan_generator.h5'
    g_model.save(model_path)
    print(f'\n✅ Generator saved to "{model_path}"')

    return history
```

```python
# ==================== BUILD & TRAIN ====================

# Latent vector dimension
LATENT_DIM = 100

# Build models
discriminator = define_discriminator()
generator     = define_generator(LATENT_DIM)
gan_model     = define_gan(generator, discriminator)

# Load & scale training data
dataset = load_real_samples()
print(f'Training data shape: {dataset.shape}  |  Range: [{dataset.min():.1f}, {dataset.max():.1f}]')

# Train!  Adjust n_epochs if you want longer / shorter runs.
history = train(
    g_model=generator,
    d_model=discriminator,
    gan_model=gan_model,
    dataset=dataset,
    latent_dim=LATENT_DIM,
    n_epochs=100,       # Maximum epochs
    n_batch=128,
    patience=5,         # Stop after 5 consecutive "fooling" epochs
    d_fake_loss_threshold=0.7
)
```

## 8. Generate 10 Fake Images with the Trained Generator

Now that training is complete, we load the saved Generator and ask it to produce **10 brand-new fake images** from random latent vectors.

```python
# Load the saved generator
saved_generator = load_model('cifar_gan_generator.h5')

# Generate 10 fake images
latent_points = generate_latent_points(LATENT_DIM, 10)
fake_images = saved_generator.predict(latent_points, verbose=0)

# Rescale from [-1, 1] to [0, 1] for display
fake_images = (fake_images + 1) / 2.0

# Plot the 10 generated images in a single row
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(fake_images[i])
    ax.set_title(f'Fake #{i+1}', fontsize=12)
    ax.axis('off')

plt.suptitle('10 Fake CIFAR-10 Images Produced by the Trained Generator', fontsize=16)
plt.tight_layout()
plt.show()
```

## 9. Training Visualization & Analysis

We plot:  
1. **Loss curves** — Discriminator (real & fake) and Generator losses over epochs  
2. **Accuracy curves** — how well the Discriminator classifies real vs fake over time  

### What to look for
- **Generator loss** should generally decrease as it learns to fool the Discriminator.  
- **Discriminator accuracy on fakes** should drop over time — meaning fakes look more realistic.  
- Ideally both losses hover around **0.6–0.7** indicating a healthy adversarial balance.

```python
epochs_ran = range(1, len(history['g_loss']) + 1)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# ---- Plot 1: Loss Curves ----
axes[0].plot(epochs_ran, history['d_loss_real'], label='D Loss (Real)', linewidth=2)
axes[0].plot(epochs_ran, history['d_loss_fake'], label='D Loss (Fake)', linewidth=2)
axes[0].plot(epochs_ran, history['g_loss'],      label='G Loss',        linewidth=2)
axes[0].axhline(y=0.7, color='red', linestyle='--', alpha=0.5, label='Early-Stop Threshold')
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Loss', fontsize=12)
axes[0].set_title('Loss Curves Over Epochs', fontsize=14)
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# ---- Plot 2: Discriminator Accuracy ----
axes[1].plot(epochs_ran, [a * 100 for a in history['d_acc_real']],
             label='D Accuracy (Real)', linewidth=2, color='tab:green')
axes[1].plot(epochs_ran, [a * 100 for a in history['d_acc_fake']],
             label='D Accuracy (Fake)', linewidth=2, color='tab:red')
axes[1].axhline(y=50, color='gray', linestyle='--', alpha=0.5, label='50% (random guess)')
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Accuracy (%)', fontsize=12)
axes[1].set_title('Discriminator Accuracy Over Epochs', fontsize=14)
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.suptitle('GAN Training Summary', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()

print('\n📊 Final epoch metrics:')
print(f"   D Loss Real : {history['d_loss_real'][-1]:.4f}")
print(f"   D Loss Fake : {history['d_loss_fake'][-1]:.4f}")
print(f"   G Loss      : {history['g_loss'][-1]:.4f}")
print(f"   D Acc  Real : {history['d_acc_real'][-1]:.2%}")
print(f"   D Acc  Fake : {history['d_acc_fake'][-1]:.2%}")
```

## 📝 Notes & Next Steps

- **This is a regular (unconditional) GAN** — it generates random CIFAR-10-like images but we cannot control *which* class is generated.  
- To generate specific classes (e.g., only horses), use a **Conditional GAN (cGAN)** that supplies a class label alongside the latent vector.  
- Training for more epochs (100–250) typically yields sharper, more recognizable images.  
- Experiment with the architecture (deeper networks, batch normalization, spectral normalization) for better results.
